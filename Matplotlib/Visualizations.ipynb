# Matplotlib example workflow

```python
# 0. import matplot lib and get it ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot Data
ax.plot(x,y)

# 4. Customize Plot
ax.set(title="Sample Plot",
       xlabel = "x-axis",
       ylabel = "y-axis");

# 5. Save and show
# fig.savefig(<PATH DIRECTORU>)
```

# Two options for subplots

```python
# Subplot 1

fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize = (10,5))

# Plot do each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(x);
```

```python
# Subplot Option 2

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(x);
```

### pyplot vs matplotlib OO
* pyplot - when you are plotting something quickly

* matplotlib OO - when you are plotting something more advanced 

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax=ax);

# ax.set_xlim([45,100])
```

```python
## OO Method from scratch

fix, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])


# Customizelabels
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel= "Cholesterol")


# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--")
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize=(10,10),
                              sharex = True);

# Add title to the figure

fig.suptitle("Heart Disease Analysis", fontsize= 15, fontweight="bold");

# ------------------------------------------------------------------------
# Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target']);

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        # xlabel="Age",
        ylabel="Cholesterol");

# ax0 legend
ax0.legend(*scatter.legend_elements(), title = "Target");


#ax0 meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# --------------------------------------------------------------------------

# Add data to ax1
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target']);
# Customize ax1
ax1.set(title="Heart Disease and thalach",
        xlabel="Age",
        ylabel="thalach");

# ax0 legend
ax1.legend(*scatter.legend_elements(), title = "Target");

#ax0 meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");



```

```python
# Set the style
plt.style.use('seaborn-whitegrid')


## OO Method from scratch

fix, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # Changes color scheme


# Customizelabels
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel= "Cholesterol")


# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--",
           color="red")
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize=(10,10),
                              sharex = True);

# Add title to the figure

fig.suptitle("Heart Disease Analysis", fontsize= 15, fontweight="bold");

# ------------------------------------------------------------------------
# Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'],
                      cmap="winter");

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        # xlabel="Age",
        ylabel="Cholesterol");

ax0.set_xlim([50,80])

# ax0 legend
ax0.legend(*scatter.legend_elements(), title = "Target");


#ax0 meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# --------------------------------------------------------------------------

# Add data to ax1
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap='winter');
# Customize ax1
ax1.set(title="Heart Disease and thalach",
        xlabel="Age",
        ylabel="thalach");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# ax0 legend
ax1.legend(*scatter.legend_elements(), title = "Target");

#ax0 meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");


```
